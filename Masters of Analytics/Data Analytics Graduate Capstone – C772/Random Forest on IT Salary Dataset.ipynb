# A. Research Question <a class="anchor" id="question"></a>

**Research Question:** Can a Random Forest model be constructed based upon the survey salary dataset?

**Hypothesis:** 

H0: A Random Forest model cannot be formulated on the survey salary dataset.
                
H1: A Random Forest can be constructed on the survey salary dataset.

The null hypothesis above attempts to verify the research question by means of stating of the condition where a Random Forest model *cannot* be created from the salary dataset. The alternative hypothesis claims that a Random Forest Model *can* be constructed on the survey salary dataset.

The study exists to create a predictive model which can predict Data Professionals Salaries in order to aid graduating students with functional and targeted market research. This study will utilize a Random Forest regression model in order to analyze the significance of particular predictor variables within the salary data survey. The predominant goals will be to identify which variables best predict the salary amounts for Data Professionals while also constructing an accurate model. According to Donges (2020), a Random Forest model is a supervised learning algorithm which merges decision trees in an ensemble type of methodology. Donges (2020) also found that utilizing a Random Forest regression model can help predict both classification and continuous variable alike. Aasim (2019) hypothesized that salary positions data can be descriptive enough to predict salaries while utilizing Random Forest. Therefore, by thoughtful and thorough understanding of the classification and discrete predictor variables, as well as their relationship between the salary, it should be possible to predict this response variable (Koehrsen, 2017).

## II. Dependent Salary Variable Analysis and Cleanup <a class="anchor" id="salary_variable"></a>

### a. Normality Test <a class="anchor" id="salary_normal"></a>

```python
# Creating X Salary value from data frame
my_data = pd.Series(df['SalaryUSD'])

# Creating Mean
mu = np.mean(my_data)

#Creating Standard Div
sigma = np.std(my_data)

# Random Sample
x=mu+sigma*np.random.randn(500)

#Specifying Bins
num_bins=30

# Adding Bin distributions
fig,ax=plt.subplots()
n,bins,patches=ax.hist(x,num_bins,density=1, facecolor='teal')

# Fitting curve over top
y=stats.norm.pdf(bins,mu,sigma)
ax.plot(bins,y,'--')

#Giving axis names, title etc.
sns.set(rc={'figure.figsize':(6,6)})
plt.xlabel("Salary", size = 13)
plt.ylabel("Probability", size=13)
plt.title("Histogram of Salary Normality", size=14)
plt.grid(True)
```

**The histogram above indicates that salary has a normal distribution.** This is found by the normal-distributive bell curve that the bins of the dataset sample fit within. The plots below are looking for outliers. An advantage of checking for the normality is a better understanding of the target variable itself. Additionally, if an analysis is parametric it would require normal distribution. While the Random Forest model can predict parametric and non-parametric alike, it is still important to discern.

### b. Outlier Analysis <a class="anchor" id="salary_outlier"></a>

```python
# Scatter plot of salary counts
sns.set(rc={'figure.figsize':(7,7)})
sns.scatterplot(x=df.index, y=my_data)
plt.xlabel("Counts", size = 13)
plt.ylabel("Salary", size=13)
plt.title("Scatter Plot of Salary Counts", fontsize=14)
plt.grid(True)
```

```python
# Violin plot of salary counts
sns.set(rc={'figure.figsize':(6,6)})
sns.violinplot(data=df, x='SalaryUSD')
plt.xlabel("Salary", size=13)
plt.title("Violin Plot of Salary Counts", fontsize=14)
plt.grid(True)
```

Based upon the violin plot and the scatter plot above, the SalaryUSD variable appears to have some extreme outliers. In the scatter plot this is indicated by the singular dots that are far from the rest. The long horizontal line indicates the same in the violin plot. More analysis on the removal of these will be performed below.

```python
# Looking at Upper Bound outliers
df.sort_values(by='SalaryUSD', ascending=False).head(5)
```

The dataset shows some outliers that are well over $250,000. It was already determined using the scatter plot that there were upper-bound outliers. A further analysis will be conducted to remove both upper outliers using a Z-Score. 

```python
# Looking at Lower Bound outliers
df.sort_values(by='SalaryUSD').head(5)
```

An initial glance of the lower Salaries indicates that there are also some lower-bound outliers. These will also be investigated further using Z-score for removal. 

```python
# Looking at Z score (Standard Deviations)
z = stats.zscore(my_data)
# Setting upper-bound Standard Deviation of 3
TopThreshold = 3
# Setting lower-bound Standard Deviation of -1.5
BottomThreshold = -1.5
#Finding outliers by Z-Score
Top_Outliers = np.where(z > TopThreshold)
Bottom_Outliers = np.where(z < BottomThreshold)
# Counter upper-bound outliers, the plus one adds the one 0
upperCount = np.count_nonzero(np.where(Top_Outliers)) + 1
print(black(r'There are %s upper-bound outliers with a standard deviation greater than three.' %upperCount, ['bold'])) 
```

```python
# Distinct Indexes for upper-bound outliers
print("Distinct Indexes for upper-bound outliers:", Top_Outliers)
```

```python
# Review upper-bound outlier example
df.loc[[556]]
```

As shown above this salary of $1,850,000 is obviously an extreme outlier for a simple data professional salary. 

```python
# Counter lower-bound outliers
bottomCount = np.count_nonzero(np.where(Bottom_Outliers)) + 1
print(black(r'There are %s lower-bound outliers with a standard deviation less than one and a half.' %bottomCount, ['bold'])) 
```

```python
# Distinct Indexes for lower-bound outliers
print("Distinct Indexes for Lower-bound outliers:", Bottom_Outliers)
```

```python
# Review lower-bound outlier example
df.loc[[1176]]
```

As shown above this salary of $135 is obviously an extreme outlier for a simple data professional salary. After scrupulous analysis on the outliers for variable SalaryUSD, outliers will be removed below for both upper and lower-bound conditions.

```python
# Counting the original rows of the dataframe
print("Number of rows in Dataframe before dropping outliers:", len(df))
totalCount = bottomCount + upperCount
print("Number of rows expected to be deleted:", totalCount)

#Removing outliers
combinedOutliers = np.unique(np.concatenate((Top_Outliers, Bottom_Outliers),axis=1))
combinedOutliers = list(combinedOutliers)
toDrop = pd.Series(index=combinedOutliers, dtype=object)
toDrop = toDrop.index.to_frame().drop_duplicates()
toDrop.columns = ['toDelete']
df = df.drop(toDrop.index)

# Counting the rows of the dataframe
print("Number of rows in Dataframe after dropping outliers:", len(df))
print("A simple audit above shows that the outliers were handled correctly.")
```

### a. Cluster Analysis <a class="anchor" id="cluster"></a>

Cluster data analysis is an unsupervised learning technique that helps to describe the data and place data into clusters or segments. Typically popular among marketing analysts in order to segment target marketing strategies, below clustering is used to group numeric data into cluster segments of the salary dataset. Due to their high correlation between YearsWithThisDatabase and YearsWithThisTypeOfJob only one variable aforementioned has been included. This will allow for a more accurately descriptive cluster model. This analysis is advantageous by trying to uncover what may not be readily acknowledged to the naked eye. However, it should also be noted that a disadvantage to this is that it can also be hard to interpret.               

```python
# Determining how many clusters for K-mean clustering
# Setting numeric only dataframe
clusterData = df[['SalaryUSD', 'YearsWithThisDatabase', 'DatabaseServers']]

# Setting distortion and clusters
distortions = []
num_clusters = range(1, 7)

# Create a list of distortions from the kmeans function
for i in num_clusters:
    cluster_centers, distortion = kmeans(clusterData.values.astype(float), i)
    distortions.append(distortion)

    
# Create a data frame with two lists - num_clusters, distortions
elbow_plot = pd.DataFrame({'num_clusters': num_clusters, 'distortions': distortions})
# Create a line plot of num_clusters and distortions
sns.lineplot(x='num_clusters', y='distortions', data = elbow_plot)
plt.xticks(num_clusters)
plt.xlabel('Number of clusters', size=13)
plt.ylabel("Inertia", size=13)
plt.title("Inertia of k-Means versus number of clusters", size=14)
plt.show()
```

By using the elbow methodology of determining how many clusters to use, it is clear that three clusters should be used in K-means, based upon the above analysis. However, K-means is particularly helpful with large datasets and is not nearly as accurate as hierarchical clustering. Thus hierarchical clustering will be use in order to continue to describe this dataset.

```python
# Format cluster data for salary and years with database
X = clusterData.iloc[:, [0, 1]].values

# Format cluster data for all numeric values in previously created dataframe
X2 = clusterData.iloc[:, [0, 2]].values

# Create hierarchical dendrogram
sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Hierarchical Clustering Dendrogram 1', size=14)
plt.xlabel('Index', size=13)
plt.ylabel('Distance', size= 13)
plt.show()
```

```python
# Create hierarchical dendrogram
sch.dendrogram(sch.linkage(X2, method='ward'))
plt.title('Hierarchical Clustering Dendrogram 2', size=14)
plt.xlabel('Index', size=13)
plt.ylabel('Distance', size= 13)
plt.show()
```

Based upon the dendrograms above and the clusters’ distances, it appears that the appropriate clusters based upon hierarchical selection is four segment clusters for YearsWithThisDatabase and SalaryUSD. For all numeric variables it appears that five clusters is appropriate based upon variable interaction.

### b. PCA (Principal Component Analysis) <a class="anchor" id="pca"></a>

PCA analysis is a great way to figure out which variables explain a majority of the variance and reduces those to what are called Principle Components. According to Jakevdp (2020), PCA is a dimensionality reduction algorithm which can help to reduce the noise of many variables. PCA is specifically advantageous in order to reduce the variables down to those that are truly important. This helps the important variables to speak with a microphone over the noise. A disadvantage is that if used incorrectly, information loss is a real risk. Below, PCA is used to determine the variance of the dataset's numeric variables.

```python
# Setting numeric only dataframe
pca_df = df[['SurveyYear', 'YearsWithThisTypeOfJob', 'YearsWithThisDatabase', 'DatabaseServers']]
# Pre processing four variables
pca_df = preprocessing.MinMaxScaler().fit_transform(pca_df)

# Execute PCA Analysis
pca = prince.PCA(
     n_components=4,
     n_iter=3,
     copy=True,
     check_input=True,
)


pca.fit(pca_df) 
pca_df
```

The code below attempts to explain for each additional variable, what the total number of variance explained by the variables is.

```python
#Principle Components in order of variance Ratio
print(f"Variance Explained by SurveyYear: {round((pca.explained_inertia_ * 100)[0], 2)}%")
print(f"Variance Explained by the addition of YearsWithThisTypeOfJob: {round((pca.explained_inertia_ * 100)[1], 2)}%")
print(f"Variance Explained by the addition of YearsWithThisDatabase: {round((pca.explained_inertia_ * 100)[2], 2)}%")
print(f"Variance Explained by the addition of DatabaseServers: {round((pca.explained_inertia_ * 100)[3], 2)}%")
```

```python
#Visual representation of the explained variance of the components
components = pd.DataFrame({'var':pca.explained_inertia_,
             'PC':['PC0','PC1','PC2','PC3']})
sns.barplot(x='PC',y="var", 
           data=components, palette="colorblind");
plt.title("Components Variance Explained", size=14)
plt.xlabel('Principal Components', size= 13)
plt.ylabel("Variance", size = 13)
plt.grid(True)
```

```python
#pca = PCA().fit(pca_df.data)
plt.plot(np.cumsum(pca.explained_inertia_))
plt.title("Components Variance Elbow Plot", size=14)
plt.xlabel('Number of Components', size=13)
plt.ylabel('Cumulative Explained Variance', size=13);
```

As seen above with the explained variance ratio, the bar graph and elbow graphs, roughly 90% (37%+25%+25%) of the variance can be explained by using the SurveyYear, YearsWithThisTypeOfJob, and columns. Because DatabaseServers only explains roughly 10% of the variance, it will be removed from the dataset in further modeling.

### b. Random Forest <a class="anchor" id="RF"></a>

According to Donges (2020), "one of the biggest advantages of random forest is its versatility. It can be used for both regression and classification tasks, and it’s also easy to view the relative importance it assigns to the input features.

Random forest is also a very handy algorithm because the default hyperparameters it uses often produce a good prediction result. Understanding the hyperparameters is pretty straightforward, and there's also not that many of them. 

One of the biggest problems in machine learning is overfitting, but most of the time this won’t happen thanks to the random forest classifier. If there are enough trees in the forest, the classifier won’t overfit the model.

The main limitation of random forest is that a large number of trees can make the algorithm too slow and ineffective for real-time predictions. In general, these algorithms are fast to train, but quite slow to create predictions once they are trained."

Additionally, the rest of the machine learning models further used in this analysis are advantageous by attempting to provide better predictive modelling and increasing the fit on the dataset. Because several variables are categorical and not strings values, a method to convert them into numeric values is required. Machine Learning models almost always require numeric variables. In order to achieve this, Hot Encoding is executed below. According to Brownlee (2020), the function transposes categorical variable into distinct features, while creating numeric flags.

```python
# Creates features to encode that are not numeric
# Manage staff implicitly has been changed to Boolean, so defining it as an object to ensure encoding
X_train['ManageStaff'] = X_train['ManageStaff'].astype('object')
features_to_encode = X_train.columns[X_train.dtypes==object].tolist()

# Printing distinct features to encode
print("Distinct features to encode:", features_to_encode)
```

The code below helps to pass over the datatypes that are hot encoded to be combined later. Also the Random Forest model classifier is defined. Additionally, a scaler variable function has been assigned in order to normalize the data of future modeling. According to GeeksforGeeks (2019), this may also help to speed up calculations as well. 

```python
# Creates encoder object
col_trans = make_column_transformer(
                        (OneHotEncoder(),features_to_encode),
                        remainder = "passthrough"
                        )

# Creates random forest classifer object
rf_classifier = RandomForestClassifier(
                      min_samples_leaf=50,
                      n_estimators=1000,
                      bootstrap=True,
                      oob_score=True,
                      n_jobs=-1,
                      random_state=seed,
                      max_features='auto'
                                       )

# Feature scaling
scaler = MaxAbsScaler ()
```

In order to prepare the pipeline of combining the newly hot coded variables and fitting the Random Forest model, selected strings in the train and test data are converted to strings.

```python
# Because random forest prefers consistent strings, changing to strings for test and train data

# Train
X_train[['SurveyYear', 'Country', 'EmploymentStatus', 'JobTitle', 'ManageStaff',
        'OtherPeopleOnYourTeam', 'Education', 'EmploymentSector', 'Gender']] = X_train[['SurveyYear', 
        'Country', 'EmploymentStatus', 'JobTitle', 'ManageStaff',
        'OtherPeopleOnYourTeam', 'Education', 'EmploymentSector', 'Gender']].astype('string')


#Test
X_test[['SurveyYear', 'Country', 'EmploymentStatus', 'JobTitle', 'ManageStaff',
        'OtherPeopleOnYourTeam', 'Education', 'EmploymentSector', 'Gender']] = X_test[['SurveyYear', 
        'Country', 'EmploymentStatus', 'JobTitle', 'ManageStaff',
        'OtherPeopleOnYourTeam', 'Education', 'EmploymentSector', 'Gender']].astype('string')


# Creates Values for X Copy to Check Actual vs Predicted Values
X_copy = X.copy()
X_copy = X_copy.fillna('na')

#Test
X_copy[['SurveyYear', 'Country', 'EmploymentStatus', 'JobTitle', 'ManageStaff',
        'OtherPeopleOnYourTeam', 'Education', 'EmploymentSector', 'Gender']] = X_copy[['SurveyYear', 
        'Country', 'EmploymentStatus', 'JobTitle', 'ManageStaff',
        'OtherPeopleOnYourTeam', 'Education', 'EmploymentSector', 'Gender']].astype('string')
```

```python
# Creates combined machine learning object
pipe = make_pipeline(col_trans, scaler, rf_classifier)

# Trains model
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

# E. Data Summary and Implications <a class="anchor" id="summary"></a> 

After methodical and scrupulous analysis, the null hypothesis can be rejected in favor of the alternative hypothesis. There is sufficient evidence to argue that a Random Forest model *can* be constructed on the survey salary dataset. However, with an accuracy in the mid-60th percentile range, there are other models that may be better fits for this salary dataset. Therefore, the recommended course of action would be to implement a more accurate model for future analysis of the data. There are two directions that could further this recommended course of action: implementation of the MLP neural network; implementation of K-Nearest Neighbor (KNN).

After comparing the ten different machine learning algorithms fitted over this dataset, the MLP neural network is by far the most accurate model at close to a mid-90th percentile accuracy score. It’s also important to reiterate that this model is very computationally expensive. To put this into perspective, this entire analysis takes roughly fifteen minutes to run, and eight minutes of that is the neural network’s algorithm (and a majority of the rest are the two cluster analyses). With that said and due to the small nature of this dataset, the neural network may be the better model performance if accuracy is the ultimate goal. 

K-Nearest Neighbor is the second most accurate model in predicting the dataset at a rounded 80% accuracy score. This model is far less computationally expensive in comparison to the neural network implementation and likely less subjected to over-fitting. If computational performance is more important than overall accuracy, KNN may be the best choice for future prediction methodology. One metric that could be used to discern which model is the best for this use case could be comparing the over-fitting and under-fitting of KNN vs. MLP. One of the packages that could be used to help explain and compare these models would be LIME.


The two most important features in this dataset are country with a value of United States and how many years with a particular database were given. The other features weren’t even close. The features could help discern even more explained variance if examined further.  


Again, it should also be stated that one of the primary limitations of this study is the fact that this dataset was constructed from a yearly survey. Survey datasets are subject to inaccurate metric or mis-keyed values and can produce some misguided conclusions. Additionally, this dataset only contains four years of survey data. Ten years of data would drastically improve the ability to predict salary using the Random Forest model. However, with ten years of data there might be enough data to predict on the SalaryUSD variable alone without having to create categorical Tiers for prediction. 

Overall, Random Forest is definitely a model that can predict salary on this dataset; however, there are other choices that can produce better, more accurate results.
