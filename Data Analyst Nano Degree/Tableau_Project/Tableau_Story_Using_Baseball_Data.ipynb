#  Tableau Story Using Baseball Data

## Summary

The following is an exploratory data analysis utilizing Tableau for visualization. The data itself contains 1,157 baseball players with such stats as their handedness (right or left), height in inches, weight in pounds, batting average, and home runs. This story explores the relationship between the handedness, weight, or height of the player and how it corresponds to the playerâ€™s batting performance. My initial presupposition is that left-handed batters, as well as taller players, correspond to better batter statistics overall.

## Design

First, I wanted to get a good feel for the top 10 batters. Since batting average has recently been under more scrutiny as a statistic, I first used home runs. I simply viewed the top 10 batters in sorted order in bar graph.  Secondly, merely to satisfy my curiosity, I viewed the same for batting average. After comparing the two lists I saw really only two names that really came to mind: Reggie Jackson (in home runs) and Pete Rose (in batting average). I decided to use two bar charts with one common color scheme for each graph. A legend wasn't particularly warranted.

Satisfying my curiosity at a high level, I decided to get more granular. I wanted to make sure that the height and weight of the data appeared in a normal distribution (bell curve). I used a standard graph known as a histogram for height and weight with a cascading color scheme. Since handedness only has three values, the distribution is most likely not as important, but I did want to see which was most common. For the handedness I used a stacked bar in order to view the distribution effectively. Upon investigation the height and weight were close to normal distributions of data, and right handedness was the most common.

I also wanted to see how each of the corresponding statistics corresponded to a batter's performance. All the graphs in this section of the story are based upon averages. The handedness performance is a standard bar graph with a custom color scheme. For weight and height, I chose a scatter plot for each average and included a trend line to show any correlation. Upon review, the left-handed batters fared better in home run performance. The weight showed a slight uptick in performance the taller a player was, and the height didn't seem to have any type of correlation to homerun average.

Lastly, I added a line chart of Average Home Runs in correlation to weight. I used an orange color for the line. This was done since weight had the most prominent relationship to Home Runs other than left handedness. It appears that 200-210 pounds gave the best result for hitting Home Runs.

**Note:** I only compared the statistics in relationship to home run count. This is largely due to my knowledge of baseball statics in general and how batting average is generally not considered a good judgment of a batter's performance. 

**Conclusion:** Due to the advantage of left-handed batters against mostly assumed right-handed pitchers, the performance of the individual batter's fared better as lefties. The weight of the batter did contribute some to the trend of home runs. The height did not have any direct relation to the average home run performance.
